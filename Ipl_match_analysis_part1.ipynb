##e. Write a query to get match_Id, Man of the match and runs scored by man of the match.

## -->merging the dataframe

```
merged_data = pd.merge(df1, df2, how='inner', on='match_id')
merged_data.columns
merged_data.head(10)
```

```
new_merged= merged_data[['match_id','batsman_runs','batsman','player_of_match']]
new_merged.groupby(['match_id','player_of_match','batsman'])['batsman_runs'].sum()
```

e.query to get match_Id, Man of the match and runs scored by man of the match

##f. Write a query to get list of team and boundary scored when the team won the toss and decided to bat first. (boundary includes 4s and 6s)

```
match_merged = merged_data[['match_id','toss_winner','toss_decision','total_runs','non_boundary']]
match_merged = match_merged[match_merged['toss_decision']=='bat']
match_merged = match_merged[(match_merged['total_runs']==4) | (match_merged['total_runs']==6)]
```

```
match_merged
```

```
match_merged['total_runs']=1
match_merged.groupby(['match_id','toss_winner','non_boundary'])['total_runs'].sum()
```

f.Total number of boundaries scored when the team won the toss and decided to bat first

##g. Write a query to return a report for Venue and City and average runs scored per match in powerplay in 1st inning. [powerplay means 1 to 6th Over including 6th Over]

```
merged_data1 = pd.merge(df1,df2, how='inner', on='match_id')
merged_data2 = pd.merge(merged_data1,df3, how='inner', on='venue_id')
```

```
venue_data=merged_data2[['venue','city','total_runs','match_id','overs']]
venue_data=venue_data[venue_data['overs']<7]
```

```
venue_data.groupby(['venue','city'])['total_runs'].sum()
```

g.report for Venue and City and average runs scored per match in powerplay in 1st inning.
