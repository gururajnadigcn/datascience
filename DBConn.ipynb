## 2.Write an object oriented program to expose the data from the database. Create a class Database
that has the following features:

```
import json
import sqlite3

class DBConnection:
    conn = ""
    
    def __init__(self, url='E:\DataGrokr\ipl_dataset\ipl.db'):
        self.url = url

    def connect(self):
        try:
            conn = sqlite3.connect(self.url)
            return conn
        except sqlite3.Error as e:
            print(e)
    
    def dict_factory(self,cursor, row):
        d = {}
        for idx, col in enumerate(cursor.description):
            d[col[0]] = row[idx]
        return d

    
    def getresult(self,tablename):
        conn.row_factory = self.dict_factory
        cursor = conn.cursor()
        self.tablename = tablename
        query = "select * from {}".format(tablename)
        cursor.execute(query)
        data = cursor.fetchall()
        print(data)
        return json.dumps(data)
```

```
DB = DBConnection('E:\DataGrokr\ipl_dataset\ipl.db')
conn = DB.connect()
```

##displaying result in JSON format

```
result = DB.getresult('iplvenue')
conn.close()
```
